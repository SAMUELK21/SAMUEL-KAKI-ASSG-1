```python
import matplotlib.pyplot as plt
#AB
x1 = [6,6]
y1 = [-3,18]
#plotting
plt.plot(x1, y1,color='Grey', label = 'AB', linewidth=3)
#BC
x2 = [6,-0.67]
y2 = [18,2]
#plotting
plt.plot(x2,y2,color='Blue',label ='BC' ,linewidth=3)
#CA
x3 = [-0.67,6]
y3 = [2,-3]
#plotting
plt.plot(x3,y3,color='Red',label = 'CA' ,linewidth=3)
#H
x4 = [2.995]
y4 = [2.993]
#plotting
plt.plot(x4,y4, color='black', label = 'H', linewidth=2)

plt.legend()
plt.annotate('A', (6,-3))
plt.annotate('B', (6,18))
plt.annotate('C',(-0.67,2))
plt.annotate ('H', (2.995,2.993))

plt.ylabel('$y-axis$')
plt.xlabel('$x-axis$')
plt.grid()
plt.savefig('Inctriangle.png',)
```
